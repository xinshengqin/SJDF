```python
plt.figure(50,figsize=(8,5),dpi=200)
sp1 = 10 # plot every this many points
sp2 = 11
sp3 = 12
sp4 = 13
sp5 = 14
sp6 = 15
skip_start = 10
time_shift = 0.0 # time shift in seconds

l1, = plt.plot( (gauges['standard'].t[::sp1]+time_shift)/3600.0, gauges['standard'].q[3,:][::sp1], \
               '-', color='b',linewidth=0.5, markersize=3,
              label='original setup, fine grid')

l2, = plt.plot( (gauges['case1'].t[::sp2]+time_shift)/3600.0, gauges['case1'].q[3,:][::sp2], \
               'd', color='c', linewidth=0.5, markersize=2,
              label='case1')

l3, = plt.plot( (gauges['case2'].t[::sp3]+time_shift)/3600.0, gauges['case2'].q[3,:][::sp3], \
               '+', color='r', linewidth=0.5, markersize=3,
              label='case2')

# l4, = plt.plot( (gauges['two_third'].t[::sp4]+time_shift)/3600.0, gauges['two_third'].q[3,:][::sp4], \
#                '--', color='k', linewidth=0.5, markersize=3,
#               label='finest, extra refine level in Discovery Bay')

l5, = plt.plot( (gauges['case3'].t[::sp5]+time_shift)/3600.0, gauges['case3'].q[3,:][::sp5], \
               'x', color='y', linewidth=0.5, markersize=3,
              label='case3')
l6, = plt.plot( (gauges['extra_refine_region'].t[::sp6]+time_shift)/3600.0, gauges['extra_refine_region'].q[3,:][::sp6], \
               '--', color='k', linewidth=0.5, markersize=3,
              label='extra_refine_region')


plt.xlim((-0.5,6))
plt.ylim((-1,1.5))
# plt.xlim((1,3))
# plt.ylim((1.5,2.5))
plt.legend(handles=[l1,l2,l3,l5,l6],loc='upper left',fontsize=12,markerscale=2.0)
plt.xlabel('Hours after earthquake',fontsize=16)
plt.ylabel(r"Water surface elevation ($m$)",fontsize=16)
plt.title('Gauge '+str(gauge_num))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.savefig('mesh_convergence_gauge_'+str(gauge_num)+'.png',dpi=400)
```
